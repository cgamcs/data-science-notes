```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import uuid
from sklearn.utils import resample
```

```python
data = pd.read_csv('clientes.csv')
data.head()
```

```python
# Anonimización
data.drop('direccion', axis=1, inplace=True)

data['edad'] = data['edad'] // 10 * 10

ruido = np.random.normal(0, 100, size=data['salario'].shape)

data['salario'] += ruido

data
```

```python
pseudo_nombre = []

for nombre in data['nombre']:
    pseudo_nombre.append(str(uuid.uuid4()))

data['nombre'] = pseudo_nombre

data
```

```python
agrupado = data.groupby('categoria')

data_balanceada = pd.DataFrame()

for nombre, grupo in agrupado:
    grupo_balanceado = resample(grupo, replace=True, n_samples=100, random_state=123)

    data_balanceada = pd.concat([data_balanceada, grupo_balanceado])

data_balanceada
```

```python
plt.figure(figsize=(10, 7))
sns.histplot(data['edad'], kde=True)
plt.title('Distribución de Edades de los Clientes con Curva de Densidad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia');
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(data, x='edad', y='salario', hue='categoria')
plt.title('Distribución del Salario segun Edad')
plt.xlabel('Edad')
plt.ylabel('Salario');
```
