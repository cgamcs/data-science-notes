```python
import pandas as pd

data = {"Id_producto": [1001, 1002, 1003, 1003],
        "Cantidad_vendida": [30, None, 25, 25],
        "Precio": [20.5, 15.0, None, 22.5]}

df = pd.DataFrame(data)
df
```

```python
df.info()
# podemos observar que de 3 columnas hay dos con datos nulos
```

Suponiendo que tenemos un DataFrame inmenso como podriamos saber cuales son las celdas nulas, se puede usar el metodo isnull, pero aun seria por observacion

```python
df.isnull()
```

Si queremos algo mas avanzado debemos usar el mismo metodo mas el metodo de suma

```python
df.isnull().sum()
```

Ya sabemos donde estan los valores faltantes, pero ahora falta el como manejar esas celdas, y esto en realidad depende del caso, habra ocasiones en las que puedas eliminar esos datos, pero en otros tal vez sean escenciales para completar la informacion

En el caso que usemos los datos faltantes, se puede hacer diferentes cosas con la informacion, se puede reemplzar por 0, por una palabra, por el promedio de los demas valores, eso depende, en realidad no hay una solucion perfecta, pero siempre hay una mejor que otra segun el objetivo

```python
# caso de elimacion
df_eliminados = df.dropna()
df_eliminados
```

```python
# caso de reemplazo
valores_nuevos = {'Cantidad_vendida': 0, 'Precio': df['Precio'].mean()}
df_rellenados = df.fillna(valores_nuevos)
df_rellenados
```

Modificacion del tipo de dato del DataFrame

```python
df_rellenados['Cantidad_vendida'] = df_rellenados['Cantidad_vendida'].astype(int)
df_rellenados
```

Procedemos a eliminar duplicados

```python
df_unicos = df_rellenados.drop_duplicates(subset='Id_producto')
df_unicos
```
