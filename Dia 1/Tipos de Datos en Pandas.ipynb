```python
import pandas as pd
```

### Suponiendo que tenemos una tabla con las columnas nombre y edad como esta:

| nombre | edad|
|--------|-----|
| Pedro  | 25  |
| Juan   | 39  |
| Lorena | 33  |

### A las tablas de datos en pandas se les dice DataFrames y a las columnas se les dice Series

```python
datos = {'nombre': ['Pedro', 'Juan', 'Lorena'], 'edad': [25, 39, 33]}
datos
```

### El metodo DataFrame sirve para crear un conjunto de datos en forma de tabla

```python
df = pd.DataFrame(datos)
df
```

### Como dijimos anteriormente pandas tiene dos tipos de datos los DataFrame y las Series

```python
type(df)
```

### Para acceder a las series de un DataFrame podemos usar la anotacion de corchetes o con punto, como se representa en el siguiente codigo:

```python
df['nombre']
```

```python
df.nombre
```

```python
type(df.nombre)
```
