### Conceptualmente podemos decir que una Serie de pandas en unidimensional que es capaz de almacenar cualquier tipo de datos

```python
serie = df['region']
serie.head()
```

### No solo se pueden crear Series a partir de DataFrames tambien puedes ser crear a partir de objetos de python como las listas usando el metodo Series de pandas

```python
datos = [10, 20, 30, 40, 50]
series2 = pd.Series(datos)
series2
```

### Algo mas que  hacer es agregarles indices personalizados a tus Series de pandas

```python
indices = ['a', 'b', 'c', 'd', 'e']
series2 = pd.Series(datos, indices)
series2
```

```python
series2['b']
```

```python
type(series2)
```

```python
type(series2['b']) # no es un integer normal, es un integer de numpy, los elementos en la serie en este caso son integers
```
