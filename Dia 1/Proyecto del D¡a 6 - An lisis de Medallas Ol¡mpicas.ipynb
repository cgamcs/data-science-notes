```python
import pandas as pd
```

```python
df = pd.read_csv('medallas.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
valores_nan = {'Oro': 0, 'Plata': 0, 'Bronce': 0}
df_completo = df.fillna(valores_nan)
df_completo

# se puede usar `df.fillna(0, inplace=True)` para reemplzar los datos en el dataframe original
```

```python
df_ordenado = df_completo.sort_values(by='Oro', ascending=False)
df_ordenado
```

```python
top_3_paises = df_ordenado.head(3)
top_3_paises
```

```python
df_mejores_paises = df_ordenado[df_ordenado['Oro'] > 10]
df_mejores_paises
```
