```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

**Sklearn** es una biblioteca que se especializa en machine learning

```python
df = pd.DataFrame()
df["Area"] = [2600,3000,3200,3600,4000]
df["Precio"] = [550000,565000,610000,680000,725000]
```

```python
plt.scatter(df['Area'], df['Precio']);
```

En este ejemplo tenemos una inmobiliaria y obtenemos un gráfico de los precios y la superficie de cada una, podemos observar que la grafica no muestra una línea progresiva en la comparación de la superficie de una casa con su precio. Si el día de mañana entra una nueva propiedad a la inmobiliaria, no podemos predecir con exactitud su precio con base en su superficie, ya que esta depende de más factores, no solo de su superficie, y ahí podemos usar machine learning con regresión lineal.

```python
X = df[['Area']]
```

Al trabajar con gráficas siempre establecemos el valor de **x** y de **y** en este caso la variable `X` está escrita con mayúscula y con doble corchete, porque con machine learning puede tener muchas columnas o series en un mismo eje, así que el eje X debe de tener un formato **bidimensional**. Por esto mismo deja de ser una serie y pasa a ser un **DataFrame** así que por convención se asigna la variable X en mayúscula.

```python
type(df[['Area']])
```

```python
type(df['Area'])
```

Ahora ya tenemos la primera variable de nuestro modelo, y esta variable es **independiente**, ya que es la **variable de entrada**, es decir, que contiene los **datos que ya conocemos**, después toca hacer la variable **dependiente**, o sea la **variable de salida**, ya que es la variable que queremos conocer. Queremos conocer el precio, según el área que le pasemos.

```python
y = df['Precio']
```

```python
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X, y)
```

```python
modelo.predict([[3300]])
```

Esta línea nos permite especular con cierto nivel de precisión cuál sería la relación entre las áreas y los precios

```python
plt.plot(df['Area'], modelo.predict(X));
```

Y en este grafico conjunto, podemos ver cuál es la cercania entre estos puntos y esta es la mejor linea posible de trasar, para que la distancia entre los puntos que ya tenemos sea la menor posible y poder hacer predicciones a futuro

```python
plt.scatter(df['Area'], df['Precio'])
plt.plot(df['Area'], modelo.predict(X));
```
