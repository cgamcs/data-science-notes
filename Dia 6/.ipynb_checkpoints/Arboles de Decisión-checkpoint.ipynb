Un árbol de decisión, a diferencia de los algoritmos de regresión, que solo retorna una probabilidad o valor el algoritmo de decisión regresa una categoría, por ejemplo para predecir si un email es importante, normal o spam, o predecir si el próximo vehículo que cruce la esquina es un camión, auto o bicicleta. El algoritmo de decisión es usado tanto para regresión como para clasificación, ya que nos permite predecir en que categoría va a caer el próximo dato que nos caiga en nuestra base de datos una vez que hayamos entrenado a nuestro modelo.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df.species.unique()
```

Para predecir que especie es la próxima flor que entre a nuestros datos, debemos entrenar a nuestro modelo con los datos del pétalo y el sépalo que tenemos en nuestro DataFrame, para poder hacer árboles de decisión también tenemos que crear conjuntos de entrenamiento y de prueba, y luego entrenar al modelo.

```python
# Variable independiente (entrenamiento)
X = df.drop('species', axis=1)
X
```

La variable dependiente es la Serie de `species`, pero como son string tenemos que convertirlos a valores numéricos con los cuales el algoritmo pueda hacer cálculos, para eso se puede usar la herramienta de **sklearn** llamada `LabelEncoder`.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
especies = le.fit_transform(df['species'])
especies
```

```python
y = especies
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
arbol = DecisionTreeClassifier()
```

```python
arbol.fit(X_entrena, y_entrena)
```

```python
plot_tree(decision_tree=arbol);
```

Ahora empezamos a estilar el árbol para poder entenderlo de mejor manera

```python
plt.figure(figsize=(20, 20))
plot_tree(decision_tree=arbol);
```

```python
plt.figure(figsize=(20, 20))
plot_tree(decision_tree=arbol, filled=True);
```

```python
plt.figure(figsize=(20, 20))
plot_tree(decision_tree=arbol, class_names=['setosa', 'versicolor', 'virginica'], filled=True);
```

```python
plt.figure(figsize=(20, 20))
plot_tree(decision_tree=arbol, class_names=['setosa', 'versicolor', 'virginica'], feature_names=df.columns.to_list(), filled=True);
```
