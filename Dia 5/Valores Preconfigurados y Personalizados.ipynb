```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
```

```python
pinguinos = pd.read_csv('penguins.csv')
pinguinos.head()
```

```python
sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g");
```

```python
sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g",
           palette='crest',
           marker='x',
           s=100);
```

```python
g = sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g",
           palette='crest',
           marker='x',
           s=100)
```

```python
g = sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g",
           palette='crest',
           marker='x',
           s=100)
g.set_axis_labels('Largo del Pico (mm)',
                 'Grosor del Pico (mm)',
                 labelpad=10);
```

```python
g = sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g",
           palette='crest',
           marker='x',
           s=100)
g.set_axis_labels('Largo del Pico (mm)',
                 'Grosor del Pico (mm)',
                 labelpad=10)
g.legend.set_title('Masa\nCorporal (g)');
```

```python
g = sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g",
           palette='crest',
           marker='X',
           s=100)
g.set_axis_labels('Largo del Pico (mm)',
                 'Grosor del Pico (mm)',
                 labelpad=10)
g.legend.set_title('Masa\nCorporal (g)')
g.figure.set_size_inches(6.5, 4.5);
```
