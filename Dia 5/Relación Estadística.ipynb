```python
import seaborn as sns
import pandas as pd
```

```python
propinas = pd.read_csv('tips.csv')
propinas
```

```python
sns.relplot(data=propinas,
          x='total_bill',
          y='tip');
```

```python
sns.relplot(data=propinas,
          x='total_bill',
          y='tip',
          kind='line');
```

```python
sns.relplot(data=propinas,
          x='total_bill',
          y='tip',
          kind='line',
          hue='smoker');
```

```python
sns.relplot(data=propinas,
          x='total_bill',
          y='tip',
          kind='line',
          hue='smoker',
          style='smoker');
```

```python
sns.relplot(data=propinas,
          x='total_bill',
          y='tip',
          hue='smoker',
          style='smoker',
          size='size');
```

```python
sns.relplot(data=propinas,
          x='total_bill',
          y='tip',
          hue='smoker',
          style='smoker',
          size='size',
          col='time');
```

```python
sns.relplot(data=propinas,
          x='total_bill',
          y='tip',
          hue='smoker',
          style='smoker',
          size='size',
          col='time',
          legend=None);
```
