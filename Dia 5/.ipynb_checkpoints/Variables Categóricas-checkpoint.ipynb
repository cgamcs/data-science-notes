```python
import seaborn as sns
import pandas as pd
```

```python
propinas = pd.read_csv("tips.csv")
propinas.head()
```

```python
sns.catplot(data=propinas,
           kind='strip',
           x='day',
           y='total_bill',
           hue='smoker');
```

```python
sns.catplot(data=propinas,
           kind='swarm',
           x='day',
           y='total_bill',
           hue='smoker');
```

```python
sns.catplot(data=propinas,
           kind='box',
           x='day',
           y='total_bill',
           hue='smoker');
```

```python
sns.catplot(data=propinas,
           kind='violin',
           x='day',
           y='total_bill',
           hue='smoker');
```

```python
sns.catplot(data=propinas,
           kind='point',
           x='day',
           y='total_bill',
           hue='smoker');
```

```python
sns.catplot(data=propinas,
           kind='bar',
           x='day',
           y='total_bill',
           hue='smoker');
```
