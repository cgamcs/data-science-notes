```python
# Importar librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Crear DataFrame
diamantes = pd.read_csv('diamonds.csv')

# Visualizar el DataFrame
diamantes.head()
```

```python
# Estadistica basica
diamantes.describe()
```

```python
#-------------------------------------
# RELACIONES ENTRE VARIABLES NUMERICAS
#-------------------------------------
sns.set(style='whitegrid')

# Creando el gráfico de relación con replot
g = sns.relplot(data=diamantes,
              x='carat',
              y='price',
              kind='scatter',
              height=6,
              aspect=2)

# Titulo
g.fig.suptitle('Relacion entre Precio y Peso en Quilates',
              va='baseline',
              ha='center')

# Labels
g.set_axis_labels('Peso en Quilates', 'Precio');
```

```python
#---------------------------------------
# DISTRIBUCION ENTRE VARIABLES NUMERICAS
#---------------------------------------
g = sns.displot(data=diamantes,
               x='price',
               kind='hist',
               kde=True,
               color='blue',
               height=4,
               aspect=2)

g.fig.suptitle('Distribucion del Precio',
              va='baseline',
              ha='center')

g.set_axis_labels('Precio', 'Frecuencia');
```

```python
#-----------------------
# COMPARACION CATEGORICA
#-----------------------
g = sns.catplot(data=diamantes,
               kind='box',
               x='cut',
               y='price',
               height=5,
               aspect=2,
               order=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])

# Titulo
g.fig.suptitle('Comparacion de Precio por la Calidad del Corte',
              va='baseline',
              ha='center')

# Labels
g.set_axis_labels('Calidad del Corte', 'Precio');
```

```python
#---------------------------------
# RELACION ENTRE 'DEPTH' Y 'PRICE'
#---------------------------------
g = sns.jointplot(data=diamantes,
                x='depth',
                y='price',
                kind='kde',
                fill=True,
                space=0,
                color='purple')

# Titulo
g.fig.suptitle('Relacion entre Profundidad y Precio',
              va='baseline',
              ha='center')

# Labels
g.set_axis_labels('Profundidad', 'Precio');
```

```python
#--------------------------------------------
# RELACIONES DE TODAS LAS VARIABLES NUMERICAS
#--------------------------------------------
g = sns.pairplot(diamantes, corner=True)

# Titulo
g.fig.suptitle('Relacion entre Variables Numericas',
              va='baseline',
              ha='center');
```
