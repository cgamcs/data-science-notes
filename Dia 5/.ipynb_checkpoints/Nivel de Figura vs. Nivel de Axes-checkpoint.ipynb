```python
import seaborn as sns
import pandas as pd
```

```python
pinguinos = pd.read_csv('penguins.csv')
pinguinos.head()
```

```python
sns.displot(data=pinguinos, # grafico nivel superior
           x='flipper_length_mm',
           hue='species',
           multiple='stack');
```

```python
sns.histplot(data=pinguinos, # grafico de nivel inferior
           x='flipper_length_mm',
           hue='species',
           multiple='stack');
```

```python
sns.displot(data=pinguinos,
           x='flipper_length_mm',
           hue='species',
           col='species');
```

Los grficos de nivel inferior no puedes hacer este tipo de estilos porque trabajan con un axis, no con una figura como los de nivel superior (relplot, displot, catplot)

```python
sns.relplot(data=pinguinos,
            x='flipper_length_mm',
            y='bill_length_mm',
            col='sex');
```

```python
graf = sns.relplot(data=pinguinos,
            x='flipper_length_mm',
            y='bill_length_mm',
            col='sex')
graf.set_axis_labels('Largo Aletas (mm)',
                    'Largo Pico (mm');
```
