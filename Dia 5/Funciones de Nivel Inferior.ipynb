```python
import seaborn as sns
import pandas as pd
```

```python
pinguinos = pd.read_csv("penguins.csv")
pinguinos.head()
```

```python
g = sns.PairGrid(pinguinos,
                hue='species')
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True);
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True);

g.map_lower(sns.kdeplot, # se escribe el grafico original pq estamos en un plot de bajo nivel (PairGrid)
           hue=None, # None pq no queremos que las distribuciones tomen un color
           level=5, # numero de capas
           color='.2'); # valor de gris
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True);

g.map_lower(sns.kdeplot,
           hue=None,
           level=5,
           color='.2');

g.map_lower(sns.scatterplot,
           marker='+');
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True);

g.map_lower(sns.kdeplot,
           hue=None,
           level=5,
           color='.2');

g.map_lower(sns.scatterplot,
           marker='+');

g.map_diag(sns.histplot,
          element='step',
          linewidth=0,
          kde=True);
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True)

g.map_lower(sns.kdeplot,
           hue=None,
           level=5,
           color='.2')

g.map_lower(sns.scatterplot,
           marker='+')

g.map_diag(sns.histplot,
          element='step',
          linewidth=0,
          kde=True)

g.add_legend(frameon=True);
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True)

g.map_lower(sns.kdeplot,
           hue=None,
           level=5,
           color='.2')

g.map_lower(sns.scatterplot,
           marker='+')

g.map_diag(sns.histplot,
          element='step',
          linewidth=0,
          kde=True)

g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6));
```
