```python
lista_grande = list(range(1000000))
array_grande = np.array(lista_grande)
```

```python
inicio_lista = time.time()
for n in lista_grande:
    n ** 2
fin_lista = time.time()

print(f'Tiempo lista: {fin_lista - inicio_lista}')
```
