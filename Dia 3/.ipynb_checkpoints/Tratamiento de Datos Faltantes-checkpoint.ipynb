```python
import numpy as np
```

```python
array = np.array([1, 2, 3, np.nan, 5, 6])
array
```

```python
np.isnan(array)
```

```python
np.mean(array)
```

```python
np.nanmean(array) # suma 1, 2, 3, 5 y 6, y los divide entre 5 ya que aunque hay 6 valores uno de ellos nan asi que no lo cuenta
```

```python
# reemplazar valores NaN con 0
array_con_0 = np.where(np.isnan(array), 0, array)
array_con_0
```

```python
# reemplazar valores NaN con el promedio
promedio = np.nanmean(array)
array_con_promedio = np.where(np.isnan(array), promedio, array)
array_con_promedio
```

```python
# eliminar valores NaN
array_filtrado = array[~np.isnan(array)]
array_filtrado
```
