```python
import numpy as np
```

```python
array1 = np.array([1, 2, 3])
array1
```

```python
array2 = np.array([[0], [10], [20], [30]])
array2
```

```python
broadcast_suma = array2 + array1
broadcast_suma
```

¿Cómo funciona la suma al ser arrays tan distintos?

Es la suma del elemento 0 del `array2` más el elemento 1 del `array1` = 1, la suma del elemento 0 del `array2` más el elemento 2 del `array1` = 2, la suma del elemento 0 del `array2` más el elemento 3 del `array1` = 3, después se pasa a la segunda fila del `array2`, la suma del elemento 10 del `array2` más el elemento 1 del `array1` = 11, la suma del elemento 10 del `array2` más el elemento 2 del `array1` = 12, la suma del elemento 10 del `array2` más el elemento 3 del `array1` = 13 y vicerversa con las démas lineas

```python
broadcast_multi = array2 * array1 # se sigue la misma corriente array2[0] * array1[0], array2[0] * array1[1], array2[0] * array1[2], ...
broadcast_multi
```
