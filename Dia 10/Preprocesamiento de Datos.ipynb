```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.datasets import load_iris
```

```python
data = np.array([[1, -1, 2], [3, 0, 0], [0, 2, -1]])
data
```

```python
scaler = MinMaxScaler()
data_escalada = scaler.fit_transform(data)
data_escalada
```

```python
iris = load_iris()
X = iris.data
X[:5]
```

```python
X_escalado = scaler.fit_transform(X)
X_escalado[:5]
```

```python
scaler2 = StandardScaler()
data_escalada2 = scaler2.fit_transform(data)
data_escalada2
```

```python
categorias = np.array([['rojo'], ['verde'], ['azul'], ['amarillo'], ['violeta'], ['naranja']])
categorias
```

```python
encoder = OneHotEncoder(sparse_output=False)
data_codificada = encoder.fit_transform(categorias)
data_codificada
```
