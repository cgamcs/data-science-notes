### K-Means Clustering o en español agrupamiento de promedios K

Este es un gráfico de dispersión y muestra a cada pingüino como se ubica según las medidas de su pico, tenemos longitud en el eje X y su profundidad en el eje Y, lo que se hizo con K-Means es pedirle que identifique 3 puntos (3 promedios) que sean el promedio de todas las demás medidas y K-Means ha encontrado las mejores ubicaciones para posicionar estos puntos (centroides), después de posicionar los centroides le ha asignado a cada punto del dataset (cada pingüino) uno de estos centroides, el más cercano que ha encontrado, a cada grupo de pingüinos le decimos clúster; significa agrupamiento, y cada clúster tiene su propio y único centroide, entonces el algoritmo K-Means sigue los siguientes pasos:

- Primero la inicialización: se elige el valor de K, es decir, el número de centroides
- Se hace una asignación: o sea cada punto de datos (cada pingüino) es asignado al centroide más cercano
- Se actualiza: se van recalculando el promedio de todos los puntos asignados en cada grupo y se reubica el centroide
- Iteración: se repiten los pasos 2 y 3 (asignación y actualización) hasta que los centroides ya no cambien significativamente

![](EjemploKMeansClustering.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(42)
data = np.random.rand(100,2)
data
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(data)
```

```python
centroides = kmeans.cluster_centers_
centroides
```

```python
etiquetas = kmeans.labels_
etiquetas
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data[:, 0], data[:, 1], c=etiquetas, cmap='viridis', marker='o')
plt.scatter(centroides[:, 0], centroides[:, 1], c='red', marker='x', s=200, label='Centroides')
plt.title('Visualización de K-Means Clustering')
plt.legend();
```

```python
pinguinos = pd.read_csv('penguins.csv')
```

```python
pinguinos.dropna(inplace=True)
```

```python
data = pinguinos[['bill_length_mm', 'bill_depth_mm']]
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)
```

```python
centroides = kmeans.cluster_centers_
etiquetas = kmeans.labels_
```

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(data=pinguinos, x='bill_length_mm', y='bill_depth_mm', hue=etiquetas, palette='viridis')
plt.scatter(centroides[:,0], centroides[:,1], c='red', s=100, marker='X', label='Centroides')
plt.title('K-Means Clustering en el conjunto de datos Pinguinos')
plt.legend()
plt.xlabel('Longitud del Pico (mm)')
plt.ylabel('Profundidad del Pico (mm');
```
