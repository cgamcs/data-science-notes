```python
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
A = np.array([[1, 2],
             [3, 4],
             [5, 6]])
```

```python
U, sigma, VT = np.linalg.svd(A)
```

```python
iris = load_iris()
X = iris.data
```

```python
X_centrado = X - np.mean(X, axis=0)
```

```python
U, sigma, VT = np.linalg.svd(X_centrado)
```

```python
k = 2
X_transformado = U[:, :k] * sigma[:k] # sirve como parte de un proceso en la descomposicion de valores de singulares
                                      # para reducir la dimensionalidad de un conjunto de datos
```

La variable **sigma** contiene los valores singulares que representan la importancia de cada componente principal en la descripción de la variabilidad de los datos, los componentes principales son nuevas direcciones en el espacio de los datos que capturan la mayor variabilidad posible.

Entonces **k** al ser igual a 2, establece que solamente vamos a usar los 2 componentes principales y solo dos componentes quiere decir que queremos ver el espacio bidimensional.

Lo que hicimos en la variable `X_transformado` fue transformar los datos originales para adecuarlos al espacio de los componentes principales que hemos seleccionado.

```python
especies = ['setosa', 'versicolor', 'virginica']
```

```python
plt.figure(figsize=(8, 6))
for i in range(0, 3):
    plt.scatter(X_transformado[iris.target == i, 0],
                X_transformado[iris.target == i, 1],
                label=especies[i])
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
plt.title('Dataset iris transformado por SVD')
```
