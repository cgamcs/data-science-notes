```python
url = 'https://api.nasa.gov/DONKI/IPS?'

def obtener_donki(fecha_inicio, fecha_fin):
    params = {
        'api_key': api_key,
        'startDate': fecha_inicio,
        'endDate': fecha_fin
    }

    respuesta = requests.get(url, params=params)

    data = respuesta.json()

    return pd.DataFrame(data)
```
