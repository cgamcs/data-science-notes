```python
import requests

url = "https://jsonplaceholder.typicode.com/posts/1"
respuesta = requests.get(url)
```

```python
type(respuesta)
```

```python
respuesta.status_code
```

```python
if respuesta.status_code == 200:
    data = respuesta.json()
    print(data)
else:
    print(f'Error {respuesta.status_code}: no disponemos de esta informacion')
```
