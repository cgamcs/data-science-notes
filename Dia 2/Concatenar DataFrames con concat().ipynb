```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Nombre': ["Juan", "Gabriela", "Elena"], 'Edad': [23, 31, 21]})
df1
```

```python
df2 = pd.DataFrame({'Nombre': ["Carmela", "Max", "Laura"], 'Edad': [34, 25, 29]})
df2
```

El emtodo conccat sirve si queremos concetenar los datos de dos tablas por ejemplo porque tienen la misma informacion pero una mas actual que la otra

```python
df_concatenado = pd.concat([df1, df2])
df_concatenado
```

```python
df_concatenado = pd.concat([df1, df2], axis=1)
df_concatenado
```

Si queremos que se reestablescan los indices podemos hacer o siguiente:

```python
df_concatenado = pd.concat([df1, df2], ignore_index=True)
df_concatenado
```

Tambien podemos diferenciar a que dataframe pertenece cada dato con keys

```python
df_concatenado = pd.concat([df1, df2], keys=['df1', 'df2'])
df_concatenado
```

```python
df_concatenado = pd.concat([df1, df2], keys=['Enero', 'Febrero'])
df_concatenado
```
