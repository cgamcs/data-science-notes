```python
import pandas as pd
```

```python
ruta = 'Top-Películas.csv'
df = pd.read_csv(ruta)
```

```python
df
```

Ordenar es la forma de tener una mejor visualizacion de los datos segun un parametro, en este caso segun la columna **rating**

```python
df_ordenados = df.sort_values(by='rating', ascending=False)
df_ordenados.head(10)
```

```python
df_ordenados = df.sort_values(by=['rating', 'recaudación(M)'], ascending=False)
df_ordenados.head(10)
```

Agrupar es reunir datos basandonos en categorías o en valores comúnes, lo cual es muy util para el analisis comparativo, En este caso queremos ver el promedio de rating segun el genero de las peliculas

```python
df_agrupado = df.groupby('género')['rating'].mean()
df_agrupado
```

```python
df_recaudacion = df.groupby('año')['recaudación(M)'].sum()
df_recaudacion.sort_values(ascending=False).head(10)
```
