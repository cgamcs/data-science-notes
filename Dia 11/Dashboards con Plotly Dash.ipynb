```python
import dash
from dash import dcc, html
```

```python
app = dash.Dash(__name__)
```

```python
app.layout = html.Div(children=[
    html.H1(children='Hola Dash'),
    html.Div(children='Este es un div'),
    dcc.Graph(id='grafico-ejemplo',
              figure={
                  'data': [
                      {'x': [1, 2, 3], 'y': [4, 1, 2], 'type': 'bar', 'name': 'Bogotá'},
                      {'x': [1, 2, 3], 'y': [4, 5, 6], 'type': 'bar', 'name': 'Buenos Aires'}
                  ],
                  'layout': {'title': 'Visualización con Dash'}
              })
])
```

```python
if __name__ == '__main__':
    app.run(debug=True)
```

```python
app = dash.Dash(__name__)
```

```python
app.layout = html.Div(children=[
    dcc.Input(id='caja-texto', type='text', value=''),
    html.Button('Enviar', id='boton'),
    html.Div(id='salida-boton', children='')
])
```

```python
if __name__ == '__main__':
    app.run(debug=True)
```

```python
@app.callback(
    dash.Output('salida-boton', 'children'),
    [dash.Input('boton', 'n_clicks')],
    [dash.State('caja-texto', 'value')]
)

def actualizar_salida(n_clicks, valor):
    if n_clicks == None:
        return 'introduzca un valor y presione enviar'
    else:
        return f'el valor ingresado es {valor}, el boton fue presionado {n_clicks} veces'
```

```python
if __name__ == '__main__':
    app.run(debug=True)
```
