```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.DataFrame({
    "etiquetas":['Manzanas', 'Bananas', 'Naranjas', 'Uvas'],
    "cantidad":[30, 60, 80, 20]
})
df
```

```python
plt.pie(df['cantidad'])
plt.show()
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'])
plt.show()
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%')
plt.show()
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%', colors=['#87ceeb', '#6495ed', '#004095', '#40e0d0'])
plt.show()
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%', colors=['#87ceeb', '#6495ed', '#004095', '#40e0d0'], shadow=True)
plt.show()
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%', colors=['#87ceeb', '#6495ed', '#004095', '#40e0d0'], shadow=True, explode=(0, 0, 0, 0.5))
plt.show()
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%', colors=['#87ceeb', '#6495ed', '#004095', '#40e0d0'], shadow=True, explode=(0, 0, 0, 0.5))
plt.title("Mis Frutas")
plt.show()
```
