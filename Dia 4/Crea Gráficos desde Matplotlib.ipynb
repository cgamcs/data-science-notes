```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Iluminacion.csv")
df
```

```python
help(plt.hexbin)
```

```python
fig, ax = plt.subplots()

ax.hexbin(df['Latitud'], df['Longitud'], df['Luminosidad'], gridsize=20)
plt.show()
```

```python
fig, ax = plt.subplots()

hb = ax.hexbin(df['Latitud'], df['Longitud'], df['Luminosidad'], gridsize=40, cmap='Greens')
ax.set_facecolor('#000000')
plt.colorbar(hb, label='Nivel de sombras')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.title('Luminosidad en Villa luz')

plt.show()
```
