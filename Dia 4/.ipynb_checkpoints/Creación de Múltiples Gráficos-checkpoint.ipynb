```python
fig, axs = plt.subplots(ncols=2, nrows=1)

ax1 = axs[0]
ax2 = axs[1]

ax1.plot([1, 2, 3, 4], [4, 3, 2, 1])

ax2.scatter([1, 2, 3, 4], [4, 3, 2, 1])

plt.show()
```

```python
fig, axs = plt.subplots(ncols=2, nrows=1)

ax1 = axs[0]
ax2 = axs[1]

ax1.plot([1, 2, 3, 4], [4, 3, 2, 1])
ax2.scatter([1, 2, 3, 4], [4, 3, 2, 1])

ax1.set_title('Grfico de Lineas')
ax2.set_title('Grfico Scatter')

plt.show()
```
