```python
import matplotlib.pyplot as plt
```

```python
numeros = [4,2,7,6,3]

plt.plot(numeros)
plt.show()
```

```python
plt.figure(facecolor="orange")
plt.plot(numeros)
plt.show()
```

```python
plt.subplots(ncols=2, nrows=2)
plt.plot(numeros)
plt.show()
```

```python
fig, axs = plt.subplots(ncols=2, nrows=2)
plt.plot(numeros)
```
