```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Crear DataFrame
df = pd.read_csv('Datos+Meteorológicos_Arg_2023.csv')
df
```

```python
# Convirtiendo fecha en string al formato correcto
print(df['Fecha'].dtype)
df['Fecha'] = pd.to_datetime(df['Fecha'], dayfirst=True,errors='coerce')

print("FIX APLIED")
print(df['Fecha'].dtype)
```

```python
# Obteniendo el mes de cada fecha
meses = []

for mes in df['Fecha']:
    meses.append(mes.month)

df['Mes'] = meses
df
```

```python
print(df['Ciudad'].unique())
ciudad = input('Introduce una ciudad: ').title()
mes = int(input("Ingrese un mes (1-12): "))

filtro = df[(df['Ciudad'] == ciudad) & (df['Mes'] == mes)]

if filtro.empty:
        print("No se encontraron resultados, revise que la ciudad sea correcta")
```
