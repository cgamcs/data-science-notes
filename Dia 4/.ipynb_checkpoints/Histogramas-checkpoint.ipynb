```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Ventas.csv")
df
```

```python
plt.hist(df['Producto'])
plt.show()
```

```python
numeros = np.random.randn(1000)
numeros
```

```python
plt.hist(numeros)
plt.show()
```

```python
plt.hist(numeros, bins=40)
plt.show()
```

```python
plt.hist(numeros, bins=40, alpha=0.5)
plt.show()
```

```python
plt.hist(numeros, bins=40, alpha=0.5, color='red')
plt.show()
```

```python
plt.hist(numeros, bins=40, alpha=0.5, color='red', edgecolor='black')
plt.show()
```

```python
help(plt.hist)
```

```python
plt.hist(numeros, bins=40, alpha=0.5, color='red', edgecolor='black', histtype='step')
plt.show()
```

```python
x1 = np.random.randn(1000)
x2 = np.random.randn(500)
x3 = np.random.randn(100)
```

```python
plt.hist(x1, histtype='stepfilled', alpha=0.3, bins=40)
plt.hist(x2, histtype='stepfilled', alpha=0.3, bins=40)
plt.hist(x3, histtype='stepfilled', alpha=0.3, bins=40)
plt.show()
```
